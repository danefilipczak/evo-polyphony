```python
from pop import Population
from pheno import Phenotype
import music21 as m21
```

```python
terminationFitness = 0
```

```python
score = m21.converter.parse("input/test2.xml")
```

```python
population = Population(score)
```

```python
fitness = float('-inf')
count = 0
generation = 0
fitnesses = []
while fitness < terminationFitness:
    population.develop()
    population.evaluate()
    population.nextGen()
    fitness = population.fittest.fitness
    generation+=1
    print(fitness, generation)
    fitnesses.append(fitness)
    population.fittest.phenotype.write("musicxml", "output/threevoice/generation" + str(generation) + ".xml")
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(fitnesses)
```
